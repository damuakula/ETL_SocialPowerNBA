```python
import pandas as pd
```

```python
nba_2017_player_wiki_file="./Resources/nba_2017_player_wikipedia.csv"
nba_2017_player_wiki_file_pd=pd.read_csv(nba_2017_player_wiki_file)
nba_2017_player_wiki_file_pd.head()
```

```python
nba_2017_wiki_df=nba_2017_player_wiki_file_pd[["names","pageviews"]]
nba_2017_wiki_df.head()
```

```python
nba_2017_wiki_df=nba_2017_wiki_df.rename(columns={"names":"Player_Name","pageviews":"Page_Views"})
nba_2017_wiki_df.head()
```

```python
nba_2017_wiki_df=nba_2017_wiki_df.groupby(['Player_Name']).sum()
nba_2017_wiki_df.head()
```

```python
nba_2017_player_salary_file="./Resources/nba_2017_salary.csv"
nba_2017_player_salary_file_pd=pd.read_csv(nba_2017_player_salary_file)
nba_2017_player_salary_file_pd.head()
```

```python
nba_2017_player_df=nba_2017_player_salary_file_pd[["NAME","TEAM"]]
nba_2017_player_df.head()
```

```python
nba_2017_player_df=nba_2017_player_df.rename(columns={"NAME":"Player_Name","TEAM":"Team"})
nba_2017_player_df.head()
```

```python
nba_2017_player_twitter_file="./Resources/nba_2017_twitter_players.csv"
nba_2017_player_twitter_file_pd=pd.read_csv(nba_2017_player_twitter_file)
nba_2017_player_twitter_file_pd.head()
```

```python
nba_2017_twitter_df=nba_2017_player_twitter_file_pd.rename(columns={"PLAYER":"Player_Name","TWITTER_FAVORITE_COUNT":"Twitter_Favorite_Count","TWITTER_RETWEET_COUNT":"Twitter_Retweet_Count"})
nba_2017_twitter_df.head()
```
