```python
import pandas as pd
```

```python
#Read in Wikipedia file
nba_2017_player_wiki_file_pd=pd.read_csv("Resources/nba_2017_player_wikipedia.csv")
nba_2017_player_wiki_file_pd.head()
```

```python
#Drop columns from Wikipedia file that is not needed and create new data frame
nba_2017_wiki_df=nba_2017_player_wiki_file_pd[["names","pageviews"]]
nba_2017_wiki_df.head()
```

```python
#Rename columns in Wikipedia data frame
nba_2017_wiki_df=nba_2017_wiki_df.rename(columns={"names":"Player_Name","pageviews":"Page_Views"})
nba_2017_wiki_df.head()
```

```python
#Group Wikipedia data frame by player and find sum of page views
nba_2017_wiki_df=nba_2017_wiki_df.groupby(['Player_Name']).sum()
nba_2017_wiki_df.head()
```

```python
#Read in Stats file
nba_2017_stats_file = pd.read_csv("./Resources/nba_2017_players_stats_combined.csv")
nba_2017_stats_file.head()
```

```python
nba_2017_stats_file.info()
```

```python
nba_2017_stats_df = nba_2017_stats_file[["PLAYER", "POSITION", "POINTS", "FG", "FGA", "FG%", "3P", "FTA", "FT%", "AST", "STL", "BLK", "TRB", "TOV"]]
nba_2017_stats_df.head()
```

```python
nba_2017_stats_df = nba_2017_stats_df.rename(columns = {"PLAYER": "Player_Name", "POSITION" : "Position", "POINTS" : "Points_PTS",\
                            "FG" :"Field_Goals_Made_FG", "FGA": "Field_Goal_Attemps_FGA",\
                            "FG%": "Field_Goal_Percentage_FG_PER", "3P" : "Three_Pointers_Made_3P" ,\
                            "FTA" : "Free_Throw_Attempts_FTA", "FT%" : "Free_Throw_Percentage_FT_PER", \
                            "AST" : "Assists_AST", "STL" : "Steals_STL", "BLK": "Blocks_BLK",\
                            "TRB" : "Rebounds_TRB", "TOV": "Turnovers_TOV"})
nba_2017_stats_df.head()
```

```python
#Calculate Effective Field Goal Percentage and add to data frame
nba_2017_stats_df["Effective_Field_Goal_Perc"] = (nba_2017_stats_df.Field_Goals_Made_FG  +(0.5*nba_2017_stats_df.Three_Pointers_Made_3P))\
/nba_2017_stats_df.Field_Goal_Attemps_FGA
```

```python
#Calculate True Shooting Percentage and add to data frame
nba_2017_stats_df["True_Shooting__Perc"] = nba_2017_stats_df.Points_PTS  / (2*(nba_2017_stats_df.Field_Goal_Attemps_FGA +(0.44*nba_2017_stats_df.Free_Throw_Attempts_FTA)))
```
