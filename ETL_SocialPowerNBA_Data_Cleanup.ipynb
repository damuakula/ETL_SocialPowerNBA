```python
import pandas as pd
from sqlalchemy import create_engine
```

# EXTRACT & TRANSFORM

```python
#Read in Wikipedia file
nba_2017_player_wiki_file_pd=pd.read_csv("Resources/nba_2017_player_wikipedia.csv")
nba_2017_player_wiki_file_pd.head()
```

```python
#Drop columns from Wikipedia file that is not needed and create new data frame
nba_2017_wiki_df=nba_2017_player_wiki_file_pd[["names","pageviews"]]
nba_2017_wiki_df.head()
```

```python
#Rename columns in Wikipedia data frame
nba_2017_wiki_df=nba_2017_wiki_df.rename(columns={"names":"Player_Name","pageviews":"Wiki_Page_Views"})
nba_2017_wiki_df.head()
```

```python
#Group Wikipedia data frame by player and find sum of page views
nba_2017_wiki_df=nba_2017_wiki_df.groupby(['Player_Name']).sum()
nba_2017_wiki_df.head()
```

```python
#Read in Twitter file
nba_2017_player_twitter_file_pd=pd.read_csv("Resources/nba_2017_twitter_players.csv")
nba_2017_player_twitter_file_pd.head()
```

```python
nba_2017_twitter_df=nba_2017_player_twitter_file_pd.rename(columns={"PLAYER":"Player_Name","TWITTER_FAVORITE_COUNT":"Twitter_Favorite_Count","TWITTER_RETWEET_COUNT":"Twitter_Retweet_Count"})
nba_2017_twitter_df.head()
```

```python
#Merge twitter and wikipedia data frame
nba_2017_social_influence_df=pd.merge(nba_2017_wiki_df,nba_2017_twitter_df,on="Player_Name",how='outer')
nba_2017_social_influence_df.head()
```

```python
#Read in Salary file
nba_2017_player_salary_file = pd.read_csv("Resources/nba_2017_salary.csv")
nba_2017_player_salary_file.head()
```

```python
#Rename columns in Salary data frame
nba_2017_salary_df = nba_2017_player_salary_file.rename(columns={"NAME": "Player_Name", "POSITION": "Position", "TEAM":"Team", "SALARY":"Salary"})
nba_2017_salary_df.head()
```

```python
#Remove duplicate players and take average of the salaries reported
nba_2017_salary_df=nba_2017_salary_df.groupby(['Player_Name','Position','Team']).mean().reset_index()
nba_2017_salary_df.head()
```

```python
#Create Player data frame from salary csv
nba_2017_player_df=nba_2017_player_salary_file[["NAME","TEAM"]]
nba_2017_player_df.head()
```

```python
#Rename column names in Player data frame
nba_2017_player_df=nba_2017_player_df.rename(columns={"NAME":"Player_Name","TEAM":"Team"})
nba_2017_player_df.head()
```

```python
#Read in Attednace file
nba_2017_attendance_file = pd.read_csv("Resources/nba_2017_attendance.csv")
nba_2017_attendance_file.head()
```

```python
#Drop columns not needed
nba_2017_attendance_df = nba_2017_attendance_file.drop(columns=['PCT'])
nba_2017_attendance_df.head()
```

```python
#Rename columns
nba_2017_attendance_df = nba_2017_attendance_df.rename(columns={"TEAM": "Team", "GMS": "Games", "TOTAL":"Total_Attendance", "AVG":"Average_Attendance"})
nba_2017_attendance_df.head()
```

```python
#Read in Team Valuations file
nba_2017_team_valuations_file = pd.read_csv("Resources/nba_2017_team_valuations.csv")
nba_2017_team_valuations_file.head()
```

```python
nba_2017_team_valuations_df = nba_2017_team_valuations_file.rename(columns={"TEAM":"Team", "VALUE_MILLIONS":"Value_Millions"})
nba_2017_team_valuations_df.head()
```

```python
nba_2017_elo_file = pd.read_csv("Resources/nba_2017_elo.csv")
nba_2017_elo_file.head()
```

```python
#Rename columns
nba_2017_elo_df = nba_2017_elo_file.rename(columns={"ELO": "ELO_Rating", "TEAM": "Team", "CONF":"Conference"})
nba_2017_elo_df.head()
```

```python
#Merge elo and team_valuation data frame
nba_2017_team_df=pd.merge(nba_2017_elo_df,nba_2017_team_valuations_df,on="Team",how='outer')
nba_2017_team_df.head()
```

```python
#Reorder columns
nba_2017_team_df=nba_2017_team_df[["Team","Conference","ELO_Rating","Value_Millions"]]
nba_2017_team_df.head()
```

```python
nba_2017_team_df=pd.merge(nba_2017_team_df,nba_2017_attendance_df,on="Team",how='outer')
nba_2017_team_df.head()
```

```python
#Calculate ELO index and add to data frame
nba_2017_team_df["ELO_Index"] = (nba_2017_team_df.ELO_Rating/nba_2017_team_df.Value_Millions)
nba_2017_team_df.head()
```

```python
#Reorder columns
nba_2017_team_df=nba_2017_team_df[["Team","Conference","ELO_Rating","ELO_Index","Value_Millions","Games","Total_Attendance","Average_Attendance"]]
nba_2017_team_df.head()
```

```python
#Read in Stats file
nba_2017_stats_file = pd.read_csv("./Resources/nba_2017_players_stats_combined.csv")
nba_2017_stats_file.head()
```

```python
nba_2017_stats_file.info()
```

```python
nba_2017_stats_df = nba_2017_stats_file[["PLAYER", "POSITION", "POINTS", "FG", "FGA", "FG%", "3P", "FTA", "FT%", "AST", "STL", "BLK", "TRB", "TOV"]]
nba_2017_stats_df.head()
```

```python
nba_2017_stats_df = nba_2017_stats_df.rename(columns = {"PLAYER": "Player_Name", "POSITION" : "Position", "POINTS" : "Points_PTS",\
                            "FG" :"Field_Goals_Made_FG", "FGA": "Field_Goal_Attemps_FGA",\
                            "FG%": "Field_Goal_Percentage_FG_PER", "3P" : "Three_Pointers_Made_3P" ,\
                            "FTA" : "Free_Throw_Attempts_FTA", "FT%" : "Free_Throw_Percentage_FT_PER", \
                            "AST" : "Assists_AST", "STL" : "Steals_STL", "BLK": "Blocks_BLK",\
                            "TRB" : "Rebounds_TRB", "TOV": "Turnovers_TOV"})
nba_2017_stats_df.head()
```

```python
#Calculate Effective Field Goal Percentage and add to data frame
nba_2017_stats_df["Effective_Field_Goal_Perc"] = (nba_2017_stats_df.Field_Goals_Made_FG  +(0.5*nba_2017_stats_df.Three_Pointers_Made_3P))\
/nba_2017_stats_df.Field_Goal_Attemps_FGA
```

```python
#Calculate True Shooting Percentage and add to data frame
nba_2017_stats_df["True_Shooting_Perc"] = nba_2017_stats_df.Points_PTS  / (2*(nba_2017_stats_df.Field_Goal_Attemps_FGA +(0.44*nba_2017_stats_df.Free_Throw_Attempts_FTA)))
```

```python
#Add data from pandas data frame to sql tables
nba_2017_player_df.to_sql(name='player_info', con=engine, if_exists='append', index=False)

nba_2017_salary_df.to_sql(name='salary', con=engine, if_exists='append', index=False)

nba_2017_stats_df.to_sql(name='stats', con=engine, if_exists='append', index=False)

nba_2017_team_df.to_sql(name='team', con=engine, if_exists='append', index=False)

nba_2017_social_influence_df.to_sql(name='social_media', con=engine, if_exists='append', index=False)
```
