```python
import pandas as pd
import os
```

```python
df = pd.read_csv("./Resources/nba_2017_players_stats_combined.csv")
df.head()
```

```python
df.info()
```

```python
df = df[["PLAYER", "POSITION", "POINTS", "FG", "FGA", "FG%", "3P", "FTA", "FT%", "AST", "STL", "BLK", "TRB", "TOV"]]
df.head()
```

```python
stat_df = df.rename(columns = {"PLAYER": "Player_Name", "POSITION" : "Position", "POINTS" : "Points_PTS",\
                            "FG" :"Field_Goals_Made_FG", "FGA": "Field_Goal_Attemps_FGA",\
                            "FG%": "Field_Goal_Percentage_FG_PER", "3P" : "Three_Pointers_Made_3P" ,\
                            "FTA" : "Free_Throw_Attempts_FTA", "FT%" : "Free_Throw_Percentage_FT_PER", \
                            "AST" : "Assists_AST", "STL" : "Steals_STL", "BLK": "Blocks_BLK",\
                            "TRB" : "Rebounds_TRB", "TOV": "Turnovers_TOV"})
```

```python
stat_df.head()
```

```python
stat_df["Effective_Field_Goal_Perc"] = (stat_df.Field_Goals_Made_FG  +(0.5*stat_df.Three_Pointers_Made_3P))\
/stat_df.Field_Goal_Attemps_FGA
```

```python
stat_df["True_Shooting__Perc"] = stat_df.Points_PTS  / (2*(stat_df.Field_Goal_Attemps_FGA +(0.44*stat_df.Free_Throw_Attempts_FTA)))
```
